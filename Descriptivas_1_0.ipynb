```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('diabetes.xlsx')
```

```python
df.head()
```

```python
categorical_columns = df.drop(columns=['X5', 'X14', 'X15', 'X16', 'Diabetes_binary'])

for column in categorical_columns:
    print("Frecuencia de valores Ãºnicos para la variable", column)
    print(df[column].value_counts())
    print()
```

```python
#BMI
plt.figure(figsize=(10, 6))
plt.hist(df['X4'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histograma de Indice de masa corporal')
plt.xlabel('Indice de masa corporal')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

GrÃ¡ficos de barras de las variables binarias segÃºn nivel de clasificaciÃ³n

```python
#HighBP
plt.figure(figsize=(10, 6))

# GrÃ¡fico de barras para la variable categÃ³rica 'HighBP' con valores encima de las barras
ax = sns.countplot(x='X1', data=df, palette='Set2')

# Configurar tÃ­tulo y etiquetas de los ejes
plt.title('DistribuciÃ³n de ClasificaciÃ³n')
plt.xlabel('TensiÃ³n arterial alta')
plt.ylabel('Frecuencia')

# Lista para almacenar los valores Ãºnicos de la clasificaciÃ³n
valores_clasificacion = []

# AÃ±adir etiquetas con las frecuencias encima de cada barra
for p in ax.patches:
    if p.get_height() not in valores_clasificacion:
        valores_clasificacion.append(p.get_height())
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

# Mostrar el grÃ¡fico
plt.show()

#Genero - Puntaje
plt.figure(figsize=(8, 4))

# GrÃ¡fico de barras para la variable categÃ³rica 'estu_genero' con valores encima de las barras
ax = sns.countplot(x='X1', hue='Diabetes_binary', data=df, palette='Set2')

# Configurar tÃ­tulo y etiquetas de los ejes
plt.title('DistribuciÃ³n de ClasificaciÃ³n por TensiÃ³n arterial')
plt.xlabel('TensiÃ³n arterial alta')
plt.ylabel('Frecuencia')

# Lista para almacenar los valores Ãºnicos de la clasificaciÃ³n
valores_clasificacion = []

# AÃ±adir etiquetas con las frecuencias encima de cada barra
for p in ax.patches:
    if p.get_height() not in valores_clasificacion:
        valores_clasificacion.append(p.get_height())
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

# Mostrar el grÃ¡fico
plt.show()
```
